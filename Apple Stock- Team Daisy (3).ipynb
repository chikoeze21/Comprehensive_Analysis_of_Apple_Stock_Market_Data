### Investigation into Outliers

#### Stock prices
Identifying outliers in these price data points can provide valuable insights into the behavior of the stock or financial instrument. Outliers in stock price data can have various meanings and implications, depending on the context and the underlying factors affecting the data.
The aim of this investigation was to identify histotical events or news which could have led to these extreme prices. These could be for example, earnings report, merger announcement, or regulatory decisions.

```python
## The dates for these outliers are obtained
outlier_indices = {}

for item in df.columns:
    # Check if the item is present as a key in the whisker_values dictionaries
    if item in whisker_values["upper"] and item in whisker_values["lower"]:
        # Filter the DataFrame to get rows where the values are outliers (either above the upper whisker or below the lower whisker)
        filtered_series = df[(df[item] > whisker_values["upper"][item]) | (df[item] < whisker_values["lower"][item])]

        # Store the indices of outlier values in a list
        outlier_indices[item] = filtered_series.index.tolist()

outlier_indices;
```

```python
import pandas as pd
from collections import defaultdict

# Create a defaultdict to store the counts of keys for each value
value_key_counts = defaultdict(int)

# Create a defaultdict to store the keys associated with each value
value_keys = defaultdict(list)

# Specify the threshold for displaying counts
threshold = 1  # Adjust this threshold as needed

# Count the number of keys for each value and store the keys
for key, value in outlier_indices.items():
    for item in value:
        value_key_counts[item] += 1
        value_keys[item].append(key)

# Create a DataFrame to display the counts and associated keys
df_counts = pd.DataFrame({'indices': list(value_key_counts.keys()),
                   'count': list(value_key_counts.values()),
                   'associated_keys': [', '.join(value_keys[item]) for item in value_key_counts.keys()]})

# Filter the DataFrame based on the threshold
df_filtered = df_counts[df_counts['count'] >= threshold]

df_filtered = df_filtered.sort_values(by="count", ascending=False)
# Here the number of features classified as outliers for each index identified are presented
df_filtered.head()
```

```python
## Now we show the frequency of these combinations
df_filtered["associated_keys"].value_counts()
```

```python
# Group by the "Associated Keys" column and aggregate the "Value" column into a list
result_df = df_filtered.groupby('associated_keys')['indices'].apply(list).reset_index()
result_df['feature_count'] = result_df['associated_keys'].str.count(',') + 1
result_df = result_df.sort_values(by="feature_count", ascending=False)
result_df
```

```python
#### Now we create a function which gets dates using the indices. This function will allow for insertion of vertical
#### lines into the timeseries analysis to identify trends associated with the selected combination.

#dff = df.copy()
def event_dates(combination_index, dff):
    combination_dates = []
    def combinations(combination_index):
        combination_dates.append(result_df.loc[combination_index,"indices"])
        return combination_dates

    dates = []
    if len(combinations(combination_index)[0]) == 1:
        dates = combinations(combination_index)[0][0]
        dates = dff.loc[dates,'date']
    else:
        dates_are = combinations(combination_index)[0]
        for item in dates_are:
            dates.append(dff.loc[item,'date'])
    return dates
```

##### In this the plot below, "Combination_index" is changed according to the index of the combinations in "result_df".
The result is a time series stock price plot with vertical lines that signigy the occurence of these outliers.

```python

# Define the start date
start_date = '2004-08-19'

# Download historical stock data for each company starting from the defined date
df_aapl = yf.download('AAPL', start=start_date, end='2023-09-05')
df_msft = yf.download('MSFT', start=start_date, end='2023-09-05')
df_goog = yf.download('GOOG', start=start_date, end='2023-09-05')
df_amzn = yf.download('AMZN', start=start_date, end='2023-09-05')

# Create a DataFrame for closing prices
df_closing_prices = pd.DataFrame({
    'Apple (AAPL)': df_aapl['Close'],
    'Microsoft (MSFT)': df_msft['Close'],
    'Google (GOOG)': df_goog['Close'],
    'Amazon (AMZN)': df_amzn['Close']
})

# Step 2: Basic Statistics
statistics = df_closing_prices.describe()

# Step 3: Visualizations
plt.figure(figsize=(12, 6))
df_closing_prices.plot(ax=plt.gca(), title='Comparison of Closing Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.grid(True)

# Step 4: Returns and Volatility
df_returns = df_closing_prices.pct_change()
annualized_returns = ((1 + df_returns.mean()) ** 252 - 1) * 100
volatility = (df_returns.std() * (252 ** 0.5)) * 100

# Step 5: Correlations
correlation_matrix = df_returns.corr()

# Print the results
print("Basic Statistics:")
print(statistics)
print("\nAnnualized Returns:")
print(annualized_returns)
print("\nVolatility (Annualized Standard Deviation of Returns):")
print(volatility)
print("\nCorrelation Matrix:")
print(correlation_matrix)

# Show the plots
plt.show()
```
